```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Gender']=le.fit_transform(df['Gender'])
df['family_history_with_overweight']=le.fit_transform(df['family_history_with_overweight'])
df['FAVC']=le.fit_transform(df['FAVC'])
df['SMOKE']=le.fit_transform(df['SMOKE'])
df['SCC']=le.fit_transform(df['SCC'])
```

```python
# Categorical columns
Y = [col for col in df.columns if df[col].dtype == 'object']
print('Categorical columns :', Y)

# Numerical columns
X = [col for col in df.columns if df[col].dtype != 'object']
print('Numerical columns :', X)

from sklearn.preprocessing import MinMaxScaler

# Initialize the MinMaxScaler
scaler = MinMaxScaler(feature_range=(0, 1))

# Extracting numerical columns
num_cols = [col for col in X if col not in Y]

# Scaling numerical columns
df[num_cols] = scaler.fit_transform(df[num_cols])
```
