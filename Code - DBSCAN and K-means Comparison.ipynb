## Format waktu ke detik

```python
# Merubah format data Waktu menjadi jumlah detik

import re
from datetime import timedelta

# Function to convert "X min Y sec" to timedelta
def parse_time_string(Waktu):
    match = re.match(r'(\d+)\s*min(?:\s*(\d+)\s*detik)?', Waktu)
    if match:
        menit = int(match.group(1))
        detik = int(match.group(2)) if match.group(2) else 0
        return timedelta(minutes=menit, seconds=detik)
    return pd.NaT  # Return NaT if the format is incorrect

# Optionally, extract time if needed in a specific format (e.g., total seconds or total minutes)
df['Waktu'] = (df['Waktu'].apply(parse_time_string)).dt.total_seconds()
```

## Transform Kolom Q

```python
#Bentuk data diubah menjadi float

columns_to_check = ["Q1","Q2", "Q3", "Q4", "Q5", "Q6", "Q7", "Q8", "Q9", "Q10", "Q11", "Q12","Q13","Q14","Q15"]

for column in columns_to_check:
    df[column] = df[column].replace("-", 0)
    df[column] = df[column].str.replace(',', '.')
    df[column] = df[column].astype(float)
    df[column].fillna(0, inplace = True)
```

#DBSCAN

```python
# DBSCAN
from sklearn.cluster import DBSCAN
from sklearn import metrics

db = DBSCAN(eps=50,min_samples=10).fit(df)
labels = db.labels_
print(labels)
core_samples_mask = np.zeros_like(db.labels_, dtype=bool)
core_samples_mask[db.core_sample_indices_] = True
```

```python
n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)
n_noise_ = list(labels).count(-1)
```

```python
print ('Estimated number of clus: %d' % n_clusters_)
print('Estimated number of noise points: %d'%n_noise_)
print('Sihouette Coefficient: %0.3f'%metrics.silhouette_score(df, labels))
```

#KMEANS

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

kmean = KMeans(n_clusters=2, random_state=0, init='k-means++', max_iter=300)
y_kmeans = kmean.fit_predict(df)

# Compute the silhouette score
score = silhouette_score(df, y_kmeans)

# Print the silhouette score
print("Silhouette Score:", score)
```
