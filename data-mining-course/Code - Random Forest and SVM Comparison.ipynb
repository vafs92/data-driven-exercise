```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Baca File
df = pd.read_csv("Pendaftar_Beasiswa_csv.csv", sep = ";")
df
```

```python
#Drop kolom yag tidak diperlukan
df = df.drop('Kabupaten',axis=1)
df = df.drop('Propinsi', axis=1)
df = df.drop('No.', axis=1)
df = df.drop('Periode', axis=1)
# df = df.drop('Angkatan', axis=1)
df
```

```python
#Mengecek Missing Value
df.isnull().sum()
```

```python
#Mengganti NaN di kolom 'Srt Ket. Tidak Mampu' dengan tidak ada
df['Srt Ket. Tidak Mampu'] = df['Srt Ket. Tidak Mampu'].replace(np.nan, "tidak ada")
df
```

```python
#Drop baris yang ada NaN
df = df.dropna()
```

```python
#Encode
# Check the unique values in the column to ensure they are "ada" and "tidak ada"
print(df['Srt Ket. Tidak Mampu'].unique())
```

```python
# Custom mapping dictionary
mapping = {'ada': 1, 'tidak ada': 0}
```

```python
# Apply the mapping to the column
df['Srt Ket. Tidak Mampu'] = df['Srt Ket. Tidak Mampu'].map(mapping)
```

```python
#Data Balancing
df2.groupby('Status beasiswa').size()
```

```python
data_1 = df2[df2['Status beasiswa'] == 'T']
data_2 = df2[df2['Status beasiswa'] == 'Y']
```

```python
from sklearn.utils import resample

data_1_resample = resample(data_1, n_samples=450, random_state=0, replace=True)
data_2_resample = resample(data_2, n_samples=450, random_state=0, replace=True)
```

```python
df2 = pd.concat([data_1_resample, data_2_resample])
```

```python
#Normalisasi
X = df2[['Angkatan','Semester', 'IPK', 'Srt Ket. Tidak Mampu', 'Penghasilan Bapak', 'Penghasilan Ibu', 'Tgh. Listrik', 'Jml. Tagunggan masih studi', 'Tunggakan', 'Keadaan Orang Tua', 'Jumlah Adik', 'Jumlah Kakak', 'Jml. Point']]
Y = df2[['Status beasiswa']]
```

```python
from sklearn.preprocessing import MinMaxScaler

# initialising the MinMaxScaler
scaler = MinMaxScaler(feature_range=(0, 1))

# Numerical columns
num_col_ = [col for col in X.columns if X[col].dtype != 'object']
x1 = X
# learning the statistical parameters for each of the data and transforming
x1[num_col_] = scaler.fit_transform(x1[num_col_])
x1.head()
```

```python
#Split Data
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=0)
```

```python
#Metrics
from sklearn.metrics import make_scorer, accuracy_score,precision_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score ,precision_score,recall_score,f1_score
```

```python
from sklearn.svm import SVC
linear_svc = SVC(max_iter=1000)
linear_svc.fit(X_train,Y_train)
Y_pred = linear_svc.predict(X_test)
accuracy_svc=round(accuracy_score(Y_test,Y_pred)* 100, 2)
acc_linear_svc = round(linear_svc.score(X_train, Y_train) * 100, 2)

```

```python
cm = confusion_matrix(Y_test, Y_pred)
accuracy = accuracy_score(Y_test,Y_pred)
precision =precision_score(Y_test, Y_pred,average='micro')
recall =  recall_score(Y_test, Y_pred,average='micro')
f1 = f1_score(Y_test,Y_pred,average='micro')
print('Confusion matrix for SVM\n',cm)
print('accuracy_SVM: %.3f' %accuracy)
print('precision_RandomForest: %.3f' %precision)
print('recall_RandomForest: %.3f' %recall)
print('f1-score_RandomForest : %.3f' %f1)
```

```python
from sklearn.ensemble import RandomForestClassifier
random_forest = RandomForestClassifier(n_estimators=100)
random_forest.fit(X_train, Y_train)
Y_prediction = random_forest.predict(X_test)
accuracy_rf=round(accuracy_score(Y_test,Y_prediction)* 100, 2)
acc_random_forest = round(random_forest.score(X_train, Y_train) * 100, 2)
```

```python
cm = confusion_matrix(Y_test, Y_prediction)
accuracy = accuracy_score(Y_test,Y_prediction)
precision =precision_score(Y_test, Y_prediction,average='micro')
recall =  recall_score(Y_test, Y_prediction,average='micro')
f1 = f1_score(Y_test,Y_prediction,average='micro')
print('Confusion matrix for DecisionTree\n',cm)
print('accuracy_DecisionTree: %.3f' %accuracy)
print('precision_DecisionTree: %.3f' %precision)
print('recall_DecisionTree: %.3f' %recall)
print('f1-score_DecisionTree : %.3f' %f1)
```

```python
#Tabel perbandingan
results = pd.DataFrame({
    'Model': [ 'Random Forest',
              'Support Vector Machine'],
    'Score': [ acc_random_forest,
              acc_linear_svc],
    "Accuracy_score":[accuracy_rf,
                      accuracy_svc
                     ]})
result_df = results.sort_values(by='Accuracy_score', ascending=False)
result_df = result_df.reset_index(drop=True)
result_df.head(9)

#Visualisasi perbandingan
plt.subplots(figsize=(12,8))
ax=sns.barplot(x='Model',y="Accuracy_score",data=result_df)
labels = (result_df["Accuracy_score"])
# add result numbers on barchart
for i, v in enumerate(labels):
    ax.text(i, v+1, str(v), horizontalalignment = 'center', size = 15, color = 'black')

ax.set_title("Komparasi Model pada Split Data 70:30", fontsize=20)
```
